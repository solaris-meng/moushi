```python
import requests
import csv

def save_appid(stype,appid):
    url = 'http://123.57.16.150/nevermore/v1/aikw/list/json/%s' % appid
    r = requests.get(url)
    
    kws = r.json()['data']

    kids = []
    for k in kws:
        kids.append(k['keyword_id'])

    for kid in kids:
        url = 'http://123.57.16.150/nevermore/v1/aikw/rsd/csv/%s/%s' % (appid, kid)
        r = requests.get(url)

        if stype =='account':
            fname = '%s.csv' % appid
            ftype = 'ab+'
        else:
            fname = '%s_%s.csv' % (appid, kid)
            ftype = 'wb+'
        with open('data/%s' % fname, ftype) as f:
            for chunk in r.iter_content(1024):
                if not chunk:
                    break
                f.write(chunk)
        print(kid)
def merge_csv(appid):
    line_one = True
    with open('data/%s_merge.csv' % appid, 'w+') as cw:
        spamwriter = csv.writer(cw, delimiter=',',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
        with open('data/%s.csv' % appid, 'r') as cr:
            spamreader = csv.reader(cr, delimiter=',', quotechar='|')
            for row in spamreader:
                if 'kid' in row:
                    if line_one:
                        line_one = False
                    else:
                        continue
                spamwriter.writerow(row)
save_appid('account','b33e77f31fe04969803d8c93eee90701')
merge_csv('b33e77f31fe04969803d8c93eee90701')
```
